
## Algunas recomendaciones formales básicas para HTML y CSS

- Esto es obligatorio en XHTML 
- Siempre que abran una etiqueta, ciérrenla.
- Para elementos sin contenidos, pueden usar "autocerrado", por ejemplo:
```html
<img src="icon.jpg"/>
```
equivale a:
```html
<img src="icon.jpg"></img>
```

- "¡No crucen las espadas!" Es decir, **no** violen los anidamientos, por ejemplo:
<h3 style="color: red; text-weight:bold">Error:<h3>
<img style="padding-left=0px; padding-right=auto;" src="https://raw.githubusercontent.com/nivaca/HD202110/main/VaughanN/img/errorhtml.png"/>

- Terminen las propiedades de CSS con `;`. Por ejemplo:
```css
a {
    font-family: serif;
    color: orange;
    }
```
```html
<a style="font-family: serif; color: orange;">...</a>
```

## Selectores de CSS

## Sintaxis
```css
selector {
  propiedad1: valores;
  propiedad2: valores;
  ...
  propiedadn: valores;
}
```

## Tipos

### [1. Con el atributo `style`]

En el HTML:
```html
<body style="background-color:powderblue;">...</body>
```

### 2. Con identificadores únicos (`id`)

En el CSS, empiezan con `#`:
```css
#mi-encabezado {
font-family: Verdana, Helvetica, sans-serif;
}
```
En el HTML:
```html
<h1 id="mi-encabezado">...</h1>
```

### 3. Con clases
En el CSS, empiezan con `.`:
```css
.encabezado-grande {
  font-size: 150%;
  font-weight: bold;
  }
```
En el HTML:
```html
<h1 class="encabezado-grande">...</h1>
<section class="encabezado-grande">...</section>
<p class="encabezado-grande">...</p>
```

### 4. Con tipos (de elementos)
En el CSS:
```css
h2 {
  background: lightpink;
  margin: 100px;
  border-color: black;
  border: 5px;
  padding: 30px;
}
```
En el HTML:
```html
<h2>...</h2>
```


### 4.1 Con tipos + clases
En el CSS:
```css
h1.encabezado-grande {
  font-size: 150%;
  font-weight: bold;
  }
```
En el HTML:
```html
<h1 class="encabezado-grande">...</h1>
<h2 class="encabezado-grande">...</h2>
<h1 class="otra">...</h1>
```

## Especificidad (jerarquía) de selectores

estilo &gt; ID &gt; clase &gt; elemento

#### Ejemplo:
CSS:
```css
h1 {
  color: red;
}

#encabezado-azul {
  color: blue;
}

.mi-encabezado {
  color: green
}
```
HTML:
```html
<h1 class="encabezado-azul" id="mi-encabezado" style="color: pink;">Bla Bla Bla</h1>
```
El texto saldrá así:
<h1 style="color: pink;">Bla Bla Bla</h1>
  

## Especificidad de las stylesheets
Según el orden en que son cargadas, por ejemplo:

En el HTML:
```html
<link rel="stylesheet" href="tres.css">
<link rel="stylesheet" href="dos.css">
<link rel="stylesheet" href="uno.css">
```
Los selectores de `uno.css` tiene mayor prevalencia que los de `dos.css` y estos que los de `tres.css`.

# Algunas propiedades comunes

## Colores
### Cómo referirse a los colores:
1. Por su nombre en CSS (cf. https://www.rapidtables.com/web/css/css-color.html)
2. Por su código hexadecimal: e.g. `#aabbcc`
3. Por su código RGB: e.g. `rgb(10, 10, 10)`
3.1. RGB y transparencia: e.g. `rgba(200, 130, 155, 0.4)`

Si buscan "color picker" en [Google](https://www.google.com/search?name=f&hl=en&q=color+picker) les aparece una herramienta muy útil.

### Propiedades de colores
- texto: `color: royalblue;`
- fondos: `background-color: #ff69b4;`
    - `background: #ff69b4;`
- bordes: `border-color: rgb(255, 200, 200);`
- gradientes: `background-image: linear-gradient(red, yellow);`

### Propiedades de fuentes
- `font-family: "Times New Roman", Times, Georgia, serif;` (si los nombres de las fuentes tiene espacio, se encierran entre comillas)
- `font-style: italic` (valores posibles: `normal, italic, oblique`)
- `font-size: 1em` (valores en unidades absolutas (e.g., `px, pc, pt, mm` etc.) o relativas (`%, em, ex, vw, wh`, etc.))

### Uso de web fonts
Es posible incorporar (gratuitamente) miles de web fonts. Por ejemplo, los de [Google](https://fonts.google.com/):

En el HTML:
```html
<head>
  <link href="https://fonts.googleapis.com/css2?family=Vollkorn&display=swap" rel="stylesheet">
</head>
```
En el CSS:
```css
body {
  font-family: 'Vollkorn', serif;
}
```

## Cajas
Todo elemento mostrado en el navegador por medio de HTML y CSS puede considerarse una caja.
![caja](https://raw.githubusercontent.com/nivaca/HD2020-20/master/img/box.png)

- `margin` es el margen o área externa a un objeto
- `border` es el borde (coloreado o no) de un objeto (puede tener muchos estilos, e.g. `border:style: dotted;`
- `padding` ("acolchado") es algo así como un margen interno, antes del borde y el margen externo

## Tamaños
También es posible precisar el ancho (`width`) y algo (`height`) de los objetos.

## La propiedad `display`
Determina cómo ha de mostrarse un elemento en la página (i.e. cómo se relaciona espacialmente con los demás objetos de su entorno)

### Valores
- `block`: ocupa todo el ancho disponible de la página
- `inline`: muestra el elemento junto con otros elementos a sus lados (en una misma línea); no es posible fijarle ancho o alto, pues lo determina automáticamente el navegador
- `inline-block`: funciona como un elemento inline, pero sí es posible fijarle ancho y alto
- `none`: oculta el elemento completamente
- ...

Para más detalles, cf. https://developer.mozilla.org/en-US/docs/Web/CSS/display
